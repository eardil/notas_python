## Construcción de DataFrames

Pandas trae un esquema a través de DataFrames, que son tablas etiquetadas e indexadas.

Se pueden construir con un diccionario:

```python
pd.DataFrame({
    'var1': [1,2,3,4],
    'var2': [10,20,30,40]
})
```

## Selección y filtrado

```python
# delitos = pd.DataFrame.from_csv('../numpy_mplot/datos/Hartnagel.csv')
plt.plot(delitos.year, delitos.mconvict)
plt.plot(delitos.year, delitos.fconvict)
plt.show()
```

Podemos acceder a las variables como característica del DataFrame o con corchetes

```python
delitos.year
delitos['year']
```

```python
delitos.rename(columns = {'year': 'anio'}).head(5)
```

```python
delitos.head(5)
```

```python
type(delitos.year)
```

Y podemos graficarlos directamente en matplotlib.

```python
plt.plot(delitos.year, delitos.mconvict)
plt.plot(delitos.year, delitos.fconvict)
plt.show()
```

Para seleccionar varias columnas usamos la lista de nombres.

```python
delitos[['year', 'mconvict']].head(5)
```

Podemos crear booleanos con condicionales sobre las columnas.

```python
delitos.year > 1960
```

```python
(delitos.year > 1960) & (delitos.year < 1967)
```

Se filtra con cualquier booleano usando corchetes.

```python
delitos[(delitos.year > 1960) & (delitos.year < 1967)]
```

Y podemos combinar las 2.

```python
delitos[delitos.year > 1960][['year', 'mconvict']]
```

## Agrupación

```python
delitos.head(5)
```

```python
delitos['decada'] = delitos.year/10
```

```python
delitos.head(5)
```

Cada columna es un objeto Series, con sus propios métodos. Podemos convertirlo a otro tipo con `astype`

```python
delitos.decada.astype(int)
```

```python
delitos.decada = delitos.decada.astype(int)
```

```python
delitos.head(5)
```

```python
delitos['prueba'] = delitos.decada * 100
```

```python
delitos.head(5)
```

```python
delitos = delitos.drop('prueba', axis = 1)
```

```python
delitos.head(5)
```

```python
delitos.groupby('decada').sum()
```
